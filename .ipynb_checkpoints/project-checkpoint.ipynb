```python
import pandas as pd
from scipy.stats import ttest_rel, ttest_ind
from scipy import stats
import matplotlib.pyplot as plt
from scipy import stats

df = pd.read_csv("lusc-rsem-fpkm-tcga-t_paired.txt", sep="\t").dropna()
df3=pd.DataFrame(df.Hugo_Symbol)
df2 = pd.read_csv("lusc-rsem-fpkm-tcga_paired.txt", sep="\t").dropna()
```

```python
# from scipy import stats
# stat,pval = stats.shapiro(df[i])
import numpy as np
L=[]
df_1=df.drop("Hugo_Symbol",axis=1)
for i in range (0,len(df)):
    stat,pval = stats.shapiro(df_1.iloc[i])
    if pval > 0.05 :
        L.append("ND")
    else:
        L.append("not")
df['paired'] = L
df
```

```python
# from scipy import stats
# stat,pval = stats.shapiro(df[i])
import numpy as np
L=[]
df_2=df2.drop("Hugo_Symbol",axis=1)
for i in range (0,len(df)):
    stat,pval = stats.shapiro(df_2.iloc[i])
    if pval > 0.05 :
        L.append("ND")
    else:
        L.append("not")
df2['paired'] = L
df2
```

```python
copy1=df[df['paired'] == 'ND']
copy2=df2[df2['paired'] == 'ND' ]
# copy2=df2[df2.Hugo_Symbol in copy1.Hugo_Symbol]
#.drop("paired",axis=1)
# copy1 = df[df.Hugo_Symbol (df2['paired'] == 'ND') and (df['paired'] == 'ND')]
s1 = pd.merge(copy1, copy2, how='inner', on=['Hugo_Symbol'])
s1
```

```python
names=np.array(s1['Hugo_Symbol'])
# x = copy1[i for i in copy1 if 
# x = copy1[i for i in copy1 if copy1 ['Hugo_Symbol'] == names]
x = copy1[copy1['Hugo_Symbol'] == name (for name in names)]
# copy1['Hugo_Symbol'] == s1['Hugo_Symbol']
```

```python
stat, p = ttest_rel(df.['paired'] == 'ND').drop("paired"), df2,axis=1)
# print('stat=%.3f, p=%.3f' % (stat, p))
L=[]
for i in range (0,len(p)):
    if p[i] > 0.05 :
        L.append("ne")
    else:
        L.append("e")
# pd.DataFrame(L)
# plt.plot(norm.pdf(p,0,2),p)
```

```python
stat, p = ttest_ind(df, df2,axis=1)
# print('stat=%.3f, p=%.3f' % (stat, p))
L1=[]
for i in range (0,len(p)):
    if p[i] > 0.05 :
        L1.append("ne")
    else:
        L1.append("e")
# pd.DataFrame(L)
```

```python
# df3=pd.DataFrame(df2)
df3["paired"]=L
df3["indp"]=L1
```
